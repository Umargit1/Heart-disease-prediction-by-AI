#Importing Libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import precision_score, recall_score
```

#Importing Dataset

```python
df = pd.read_csv("/content/heart.csv")
df.head(5)
```

#Building Dataset
1. Dropping unnecessary Columns
2. Performing One-hot encoding

```python
df = df.drop(columns=['Sex', 'ExerciseAngina'])
df = pd.get_dummies(df, columns=['ChestPainType', 'RestingECG', 'ST_Slope'])
df.head(5)
```

#Test/Train Split

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop('HeartDisease', axis=1), df['HeartDisease'], test_size=0.2, random_state=42)
```

#Fitting Model On Training Data

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

#Testing Model On Test Data

```python
y_pred = lr.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Precision: ", precision)
print("Recall: ", recall)
```
